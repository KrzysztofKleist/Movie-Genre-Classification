```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def get_genre_distribution(csv) -> dict:
    df = pd.read_csv(csv, sep="\t")
    genres_count = {}

    for genre in pd.DataFrame(df.columns).loc[3:, 0].values.tolist():
        genres_count[genre] = 0

    # Loop over subfolders in the parent folder
    for index, row in df.iterrows():
        genres_count[row["genre"]] += 1

    genres_count = dict(sorted(genres_count.items(), key=lambda x: x[1], reverse=True))
    
    plt.figure(figsize = (4, 4))

    plt.bar(genres_count.keys(), genres_count.values())
    
    plt.xlabel("Genres")
    plt.ylabel("Occurrence")
    plt.title("Occurence of genres\n{}".format(csv))
    

    return genres_count
```

```python
df_train_random = pd.read_csv("data_train_random_aug_all.tsv", sep="\t")
df_test_random = pd.read_csv("data_test_random.tsv", sep="\t")
df_train_separate_movies = pd.read_csv("data_train_separate_movies_aug_all.tsv", sep="\t")
df_test_separate_movies = pd.read_csv("data_test_separate_movies.tsv", sep="\t")
```
